## Begin Analysis

```python
# Select alll providers with more than 50% of patients with heart failure

providers_DC = providers_DC.loc[providers_DC['perc_of_benefs_w_heart_failure'] > 50]
providers_DC['state'].value_counts()
```

```python
# Select List values from greatest to least

top_heart_failure_providers = providers_DC.sort_values('perc_of_benefs_w_heart_failure', ascending=False)
top_heart_failure_providers.head()
```

```python
# Find Averages of cost totals, add to provider

charge_avg = top_heart_failure_providers['charge_total'] / top_heart_failure_providers['services_count']
allowed_avg = top_heart_failure_providers['allowed_total'] / top_heart_failure_providers['services_count']
payment_avg = top_heart_failure_providers['payment_total'] / top_heart_failure_providers['services_count']
standardized_avg = top_heart_failure_providers['total_standardized_amount'] / top_heart_failure_providers['services_count']

for index, row in top_heart_failure_providers.iterrows():
    charge_avg = top_heart_failure_providers['charge_total'] / top_heart_failure_providers['services_count']
    allowed_avg = top_heart_failure_providers['allowed_total'] / top_heart_failure_providers['services_count']
    payment_avg = top_heart_failure_providers['payment_total'] / top_heart_failure_providers['services_count']
    standardized_avg = top_heart_failure_providers['total_standardized_amount'] / top_heart_failure_providers['services_count']
    
    top_heart_failure_providers['Charge Average'] = round(charge_avg, 2)
    top_heart_failure_providers['Allowed Average'] = round(allowed_avg, 2)
    top_heart_failure_providers['Payment Average'] = round(payment_avg, 2)
    top_heart_failure_providers['Standardized Average'] = round(standardized_avg, 2)
    

top_heart_failure_providers.head()
```

```python
# Drop outliers

top_heart_failure_providers = top_heart_failure_providers.loc[top_heart_failure_providers['services_count'] < 5000]

types = top_heart_failure_providers['type']
types.value_counts()
```

```python
top_heart_failure_providers.head()
```

```python
# Clean up header names for graph

low_cost_providers = top_heart_failure_providers[['provider_id', 'last_name', 'first_name', 'middle_initial', 'credentials', 'address_1', 'address_2', 'city', 'state', 'type', 'hcpcs_count', 'services_count', 'perc_of_benefs_w_heart_failure', 'avg_risk_score', 'Charge Average']]
low_cost_providers = low_cost_providers.rename(columns = {'provider_id':'Provider ID', 
                                                                'last_name':'Last Name',
                                                                'first_name':'First Name', 
                                                                'middle_initial': 'Middle Initial',
                                                                'credentials': 'Credentials',
                                                                'address_1':'Address 1', 
                                                                'address_2':'Address 2', 
                                                                'city':'City', 
                                                                'state':'State', 
                                                                'type':'Type',
                                                                'hcpcs_count':'HCPCS Count',
                                                                'services_count':'Service Count', 
                                                                'perc_of_benefs_w_heart_failure':'Percent of Patients with Heart Failure', 
                                                               'avg_risk_score':'Average Risk Score'})



low_cost_providers['Provider ID'].count()
```

```python
# Select lower average charges and great service count for best low cost providers

low_cost_providers_1 = low_cost_providers.loc[low_cost_providers['Charge Average'] < 500]
low_cost_providers_2 = low_cost_providers_1.loc[low_cost_providers['Service Count'] > 300]
low_cost_providers_3 = low_cost_providers_2.loc[low_cost_providers['HCPCS Count'] > 20]

low_cost_providers = low_cost_providers_3.sort_values('Charge Average', ascending=True)
low_cost_providers = low_cost_providers.reset_index(drop=True)

low_cost_providers
```
